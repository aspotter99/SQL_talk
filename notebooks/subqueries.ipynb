```python
#import sqlite3 and pandas
import pandas as pd
import sqlite3 as sl

#connecct to database object
conn = sl.connect('NYT_covid.db')
c = conn.cursor()
```

```python
#create new database table for election data
election = pd.read_csv('https://raw.githubusercontent.com/MEDSL/2018-elections-unoffical/master/election-context-2018.csv', index_col=0, encoding='latin-1')

c.execute('DROP TABLE IF EXISTS election')
c.execute('CREATE TABLE election (state str, county str, fips int, trump16 int, clinton16 int, otherpres16 int, romney12 int, obama12 int, otherpres12 int, demsen16 int, repsen16 int, othersen16 int, demhouse16 int, rephouse16 int, otherhouse16 int, demgov16 int,repgov16 int, othergov16 int, repgov14 int, demgov14 int, othergov14 int,total_population int, cvap float, white_pct float,  black_pct float, hispanic_pct float, nonwhite_pct float, foreignborn_pct float, female_pct float, age29andunder_pct float, age65andolder_pct float, median_hh_inc float, clf_unemploy_pct float, lesshs_pct float,lesscollege_pct float, lesshs_whites_pct float, lesscollege_whites_pct float, rural_pct float, ruralurban_cc float)')
conn.commit()

election.to_sql('election',conn, if_exists='replace')
```

```python
df = pd.read_sql('''SELECT c.state, c.county, cases, deaths, cases / deaths as 'case/death ratio', ALWAYS
                    FROM counties c
                    JOIN mask_use ON c.fips = countyfp
                    JOIN election e ON e.fips = c.fips
                    WHERE c.date == '2020-09-07'
                    ORDER BY clf_unemploy_pct DESC
                    LIMIT 25''',conn)
df
```

```python
df = pd.read_sql('''SELECT c.state, c.county, c.cases, c.deaths, c.cases/c.deaths as 'case/death ratio', m.ALWAYS
                    FROM counties as c
                    JOIN mask_use as m ON c.fips = m.countyfp
                    JOIN election as e on c.fips = e.fips
                    WHERE date == '2020-09-07'
                    ORDER BY e.clf_unemploy_pct DESC
                    LIMIT 25''',conn)
df
```
