```python
import pandas as pd
import sqlite3 as sl

#connecct to database object
conn = sl.connect('NYT_covid.db')
c = conn.cursor()
```

```python
d = pd.read_sql('''SELECT COUNT (county)
                    FROM counties
                    WHERE deaths >= '500'
                    AND date == '2020-09-07'
                    ''',conn)
d           
```

```python
d = pd.read_sql('''SELECT COUNT (county)
                    , MIN (deaths)
                    ,MAX (deaths)
                    ,AVG (deaths)
                    ,SUM (deaths)
                    FROM counties
                    WHERE deaths >= '500'
                    AND date == '2020-09-07'
                    ''',conn)
d
```

```python
d = pd.read_sql('''SELECT state
                   ,COUNT (county)
                    , MIN (deaths)
                    ,MAX (deaths)
                    ,AVG (deaths)
                    ,SUM (deaths)
                    FROM counties
                    WHERE deaths >= '500'
                    AND date == '2020-09-07'
                    GROUP BY state
                    ''',conn)
d
```

```python
d = pd.read_sql('''SELECT state
                   ,COUNT (county)
                    , MIN (deaths)
                    ,MAX (deaths)
                    ,AVG (deaths)
                    FROM counties
                    WHERE deaths >= '500'
                    AND date == '2020-09-07'
                    GROUP BY state
                    HAVING COUNT (county) >= 5
                    ''',conn)
d
```

```python
d = pd.read_sql('''SELECT state
                   ,COUNT (county)
                    , MIN (deaths)
                    ,MAX (deaths)
                    ,AVG (deaths)
                    ,AVG (ALWAYS)
                    FROM counties
                    JOIN mask_use ON fips = countyfp
                    WHERE deaths >= '500'
                    AND date == '2020-09-07'
                    GROUP BY state
                    HAVING COUNT (county) >= 5
                    ''',conn)
d
```
