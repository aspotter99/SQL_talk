```python
import pandas as pd
import sqlite3 as sl
```

```python
#create database object
conn = sl.connect('NYT_covid.db')
c = conn.cursor()
```

```python
# this will give you the names of the tables that were created and are currently in the db.
res = conn.execute("SELECT name FROM sqlite_master WHERE type='table';")
for name in res:
    print (name[0])
```

```python
# SQL to return all rows, all columns of the counties table
df = pd.read_sql('''select * from counties''',conn)
df
```

```python
# SQL to return all rows with only the state and cases columns
df = pd.read_sql('''SELECT state, cases FROM counties''',conn)
df
```

```python
# SQL to return unique values in the state column
df = pd.read_sql('''select distinct state from counties''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county from counties order by county''',conn)
df
```

```python
df = pd.read_sql('''select distinct county from counties''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county from counties where state == 'Texas' order by county''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county from counties where deaths == '10' order by state''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county, deaths from counties 
                    where state == 'Texas' and deaths == '10' order by state, county''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county, deaths 
                    from counties
                    where state == 'Texas' and (deaths <= '5' or deaths >= '50')
                    order by deaths, state, county''',conn)
df
```

```python
df = pd.read_sql('''select *
                    from counties
                    where state == 'Texas' and (deaths <= '5' or deaths >= '50')
                    order by deaths, state, county''',conn)
df
```

```python
df = pd.read_sql('''select * 
                    from counties 
                    where state == 'Texas' 
                        and deaths >= '500' 
                        and date = '2020-09-07' ''',conn)
df
```

```python
df = pd.read_sql('''select * from mask_use''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county, fips as c_countyfp, countyfp as m_countyfp, frequently, always
                    from counties
                    join mask_use on fips = countyfp
                    order by fips''',conn)
df
```

```python
df = pd.read_sql('''select distinct state, county, frequently, always
                    from counties
                    join mask_use on fips = countyfp
                    order by fips''',conn)
df
```

```python
df = pd.read_sql('''SELECT date, state, county, deaths, FREQUENTLY, ALWAYS
                    FROM counties
                    JOIN mask_use on fips = countyfp
                    WHERE state == 'Texas'
                    AND deaths >= '500'
                    AND date == '2020-09-07'
                    order by state''',conn)

df              
```
